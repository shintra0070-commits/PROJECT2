```python
'''
일반적으로 다중공선성(VIF ≥ 10)이 존재하는 변수들은 서로 높은 상관관계를 가지므로 제거하는 것이 권장된다. 
그러나 본 연구에서는 시계열 특성을 반영한 Lag 기반 예측 모델을 구축하였으며, 
변수들의 시간적 의존성과 경제적 중요성을 고려하여 해당 변수들을 제거하지 않았다.

따라서 회귀모델을 사용하여 분석 진행을 해보겠다.


이에 따라 다중공선성으로 인한 계수 불안정 문제를 완화하기 위해 정규화 기법이 적용된 Ridge 회귀모형을 활용하였으며,
동시에 비선형 관계를 효과적으로 반영할 수 있는 트리 기반 모델인 Random Forest 및 XGBoost를 함께 적용하여 예측 성능을 비교 분석하였다.

또한, 각 모델의 예측 성능은 R², MAE, RMSE 지표를 활용하여 비교 평가하였다.
'''
```

```python
### 시각화를 위한 라이브러리들 중에 기본 라이브러리
import matplotlib.pyplot as plt
import seaborn as sns

### 한글 깨짐 방지 처리
plt.rc("font", family="Malgun Gothic")
# Mac
# plt.rc("font", family="AppleGothic")

### 마이너스 특수기호 깨징 방지 처리
plt.rcParams["axes.unicode_minus"] = False
import pandas as pd
import numpy as np

### 상관관계 검정 라이브러리
from scipy.stats import spearmanr
from scipy.stats import pearsonr

### 데이터 분리 라이브러리
from sklearn.model_selection import train_test_split

### 가중치 라이브러리
# 특성 증가 라이브러리
from sklearn.preprocessing import PolynomialFeatures
# 정규화 라이브러리
from sklearn.preprocessing import StandardScaler ,MinMaxScaler

### 모델 라이브러리
# KNN 회귀모델 라이브러리
from sklearn.neighbors import KNeighborsRegressor
# 선형회귀모델 라이브러리
from sklearn.linear_model import LinearRegression
# 다른 모델 적용 : 릿지, 라쏘
from sklearn.linear_model import Ridge, Lasso
# 앙상블 모델 - 랜덤포레스트
from sklearn.ensemble import RandomForestRegressor
# 앙상블 모델 - 엑스트라트리
from sklearn.ensemble import ExtraTreesRegressor
# 앙상블 모델 - 그레디언트부스팅
from sklearn.ensemble import GradientBoostingRegressor
# 앙상블 모델 - 히스트그레디언트부스팅
from sklearn.ensemble import HistGradientBoostingRegressor
# 앙상블 모델 - 엑스지부스트
from xgboost import XGBRegressor


### 평가에 사용할 라이브러리
# <평가 결과를 이용한 모델 우선순위>
# - r2_score > (mse > mae)

# 평균절대오차 
from sklearn.metrics import mean_absolute_error
# 평균제곱오차
from sklearn.metrics import mean_squared_error
# 결정계수(R2)
from sklearn.metrics import r2_score

### 하이퍼파라메터 튜닝 클래스(모델)
from sklearn.model_selection import GridSearchCV

### vif수치 확인
from statsmodels.stats.outliers_influence import variance_inflation_factor

from functools import reduce
from sklearn.pipeline import Pipeline
```

```python
### 배운 모델은 다 돌려보기
X = final_df[["Brent oil", "철광석가격", "석탄가격", "GPR"]]
y = final_df["BDI"]

trained_models = {}

# 1. 시계열 split
# =========================
split = int(len(X) * 0.8)

X_train = X.iloc[:split]
X_test = X.iloc[split:]

y_train = y.iloc[:split]
y_test = y.iloc[split:]

dates_test = final_df["date"].iloc[split:]

# =========================
# 2. 모델 정의
# =========================
models = {

    # 선형
    "Linear": Pipeline([
        ("scaler", StandardScaler()),
        ("model", LinearRegression())
    ]),
    
    "Ridge": Pipeline([
        ("scaler", StandardScaler()),
        ("model", Ridge(alpha=1.0))
    ]),
    
    "Lasso": Pipeline([
        ("scaler", StandardScaler()),
        ("model", Lasso(alpha=0.01))
    ]),

    # 트리
    "RandomForest": RandomForestRegressor(n_estimators=200, random_state=42),
    "ExtraTrees": ExtraTreesRegressor(n_estimators=200, random_state=42),

    # 부스팅
    "GradientBoosting": GradientBoostingRegressor(random_state=42),
    "HistGradientBoosting": HistGradientBoostingRegressor(random_state=42),
    "XGBOOST" : XGBRegressor(random_state=42)
}

# =========================
# 3. 결과 저장
# =========================
results = []

# MAPE 함수
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# =========================
# 4. 모델 학습 및 평가
# =========================
for name, model in models.items():
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)
    mape_val = mape(y_test, y_pred)

    results.append([name, r2, rmse, mae, mape_val])

    trained_models[name] = model  
    
    y_pred = model.predict(X_test)

    
    
    
    
    print("\n===== {name} =====")
    print(f"R2   : {r2:.4f}")
    print(f"RMSE : {rmse:.4f}")
    print(f"MAE  : {mae:.4f}")
    print(f"MAPE : {mape_val:.2f}%")

    # ===== 시각화 =====
    plt.figure(figsize=(14,6))
    plt.plot(dates_test, y_test.values, label="Actual BDI")
    plt.plot(dates_test, y_pred, label=f"{name} Predicted")
    plt.legend()
    plt.title(f"{name} Prediction")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
            
# =========================
# 5. 결과 테이블
# =========================
results_df = pd.DataFrame(results, columns=["Model", "R2", "RMSE", "MAE", "MAPE"])

# 정렬 (R2 기준)
results_df = results_df.sort_values(by="R2", ascending=False).reset_index(drop=True)

print("\n=== 전체 모델 비교 ===")
print(results_df)

# =========================
# 6. BEST 모델 자동 선택
# =========================
best_model = results_df.iloc[0]

print("\n🔥 추천 모델")



print(best_model)
```

```python
"""
현재 독립변수는 "Brent oil", "철광석가격", "석탄가격", "GPR" 로 종속변수인 BDI를 예측하는 모델이지만 
이렇게 독립변수를 넣으면 예측결과가 너무 안좋은 모습을 보이고 있다.

결과가 안좋은 이유
 1. 데이터의 양이 부족하다.
 2. 시계열 데이터인데 과거 종속변수 데이터를 넣어주지 않았다.

이렇게 두가지의 이유를 파악해 하나씩 해결 해보겠다
"""
```

```python
### 과거 데이터 컬럼 추가
for lag in [1,2,3]:
    final_df[f"BDI_lag{lag}"] = final_df["BDI"].shift(lag)
    
for lag in [1,2,3]:
    final_df[f"Brent_lag{lag}"] = final_df["Brent oil"].shift(lag)
    final_df[f"iron_lag{lag}"] = final_df["철광석가격"].shift(lag)

for lag in [1,2,3]:
    final_df[f"coal_lag{lag}"] =final_df["석탄가격"].shift(lag)
    final_df[f"GPRD_lag{lag}"] = final_df["GPR"].shift(lag)

### 종속변수와 독립변수 시차 맞추기
final_df["target"] = final_df["BDI"].shift(-1)

final_df = final_df.dropna().reset_index(drop=True)

### 데이터 분리
X = final_df[
    [
        "BDI_lag1", "BDI_lag2", "BDI_lag3",
        "Brent_lag1", "Brent_lag2", "Brent_lag3",
        "iron_lag1", "iron_lag2", "iron_lag3",
        "coal_lag1", "coal_lag2", "coal_lag3",
        "GPRD_lag1", "GPRD_lag2", "GPRD_lag3"
    ]
]

y = final_df["target"]

trained_models = {}

# =========================
# 0. y 로그 변환
# =========================
y_log = np.log1p(y)

# =========================
# 1. 시계열 split
# =========================
split = int(len(X) * 0.8)

X_train = X.iloc[:split]
X_test = X.iloc[split:]

y_train = y_log.iloc[:split]
y_test = y_log.iloc[split:]

# 👉 실제값 (평가용)
y_test_real = y.iloc[split:]
dates_test = final_df["date"].iloc[split:]

# =========================
# 2. 스케일링 (for문 전에)
# =========================
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =========================
# 3. 모델 정의
# =========================
models = {
    "Linear": LinearRegression(),
    "Ridge": Ridge(alpha=1.0),
    "Lasso": Lasso(alpha=0.01),

    "RandomForest": RandomForestRegressor(n_estimators=200, random_state=42),
    "ExtraTrees": ExtraTreesRegressor(n_estimators=200, random_state=42),

    "GradientBoosting": GradientBoostingRegressor(random_state=42),
    "HistGradientBoosting": HistGradientBoostingRegressor(random_state=42),

    "XGBOOST": XGBRegressor(random_state=42)
}

# =========================
# 4. 평가 함수
# =========================
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# =========================
# 5. 모델 학습 및 평가
# =========================
results = []

for name, model in models.items():

    # 학습
    model.fit(X_train_scaled, y_train)

    # 로그 상태 예측
    y_pred_log = model.predict(X_test_scaled)

    # 역로그 변환
    y_pred_real = np.expm1(y_pred_log)

    # 평가 (실제값 기준)
    r2 = r2_score(y_test_real, y_pred_real)
    rmse = np.sqrt(mean_squared_error(y_test_real, y_pred_real))
    mae = mean_absolute_error(y_test_real, y_pred_real)
    mape_val = mape(y_test_real, y_pred_real)

    results.append([name, r2, rmse, mae, mape_val])
    trained_models[name] = model

    print(f"\n===== {name} =====")
    print(f"R2   : {r2:.4f}")
    print(f"RMSE : {rmse:.4f}")
    print(f"MAE  : {mae:.4f}")
    print(f"MAPE : {mape_val:.2f}%")

    # =========================
    # 시각화
    # =========================
    plt.figure(figsize=(14,6))
    plt.plot(dates_test, y_test_real.values, label="Actual BDI")
    plt.plot(dates_test, y_pred_real, label=f"{name} Predicted")
    plt.legend()
    plt.title(f"{name} Prediction (Log → Inverse)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# =========================
# 6. 결과 정리
# =========================
results_df = pd.DataFrame(results, columns=["Model", "R2", "RMSE", "MAE", "MAPE"])
results_df = results_df.sort_values(by="R2", ascending=False).reset_index(drop=True)

print("\n=== 전체 모델 비교 ===")
print(results_df)

# =========================
# 7. BEST 모델 선택
# =========================
best_model = results_df.iloc[0]

print("\n 추천 모델")
print(best_model)
```

```python
"""
독립변수에 종속변수의 과거 데이터를 시차를 나눠서 (1일전, 2일전 3일전 등등) 독립변수에 넣었더니
몇몇 모델들은 R2의 값이 음수를 탈출해 양수의 값을 보이기 시작했다.

하지만 여전히 쓸 수 없는 모델이였다. 이유는 훈련데이터와 테스트데이터가 너무 작기때문이다

지금 모델에 쓰는 독립변수들의 데이터를 늘리기 위해서 BDI수치를 월별 평균으로 예측하는것이 아니라
내일의 BDI수치를 예측하는 방향으로 틀기로 한다.
따라서 독립변수들도 월 데이터가 아닌 일 데이터로 다시 수집해서 모델 훈련을 시작하겠다.
"""
```

```python
coal_daily  = pd.read_csv("./일별데이터/석탄가격(일별).csv")
iron_daily  = pd.read_csv("./일별데이터/철광석가격(일별).csv")
bdi_daily   = pd.read_csv("./일별데이터/BDI(일별).csv")
gpr_daily   = pd.read_csv("./일별데이터/지정학적리스크(일별).csv")
brent_daily = pd.read_csv("./일별데이터/브렌트유가격(일별).csv")
```

```python
#  리스트 생성
df_list = [brent_daily, iron_daily, gpr_daily, bdi_daily, coal_daily]

#  날짜 범위 설정
start_date = "2000-01"
end_date = "2025-01"

#  df_list 안 모든 데이터프레임에 필터 적용
df_list = [df_daily[(df_daily["Date"] >= start_date) & (df_daily["Date"] <= end_date)].reset_index(drop=True) for df_daily in df_list]


#  한 번에 merge
df_temp = df_list[0]

for df in df_list[1:]:
    df_temp = pd.merge(df_temp, df, on="Date", how="inner")

df_daily = df_temp
```

```python
# =========================
# 1. Lag Feature 생성
# =========================
for lag in [1,2,3,5,10]:
    df_daily[f"BDI_lag{lag}"] = df_daily["BDI"].shift(lag)
    
for lag in [1,3,5]:
    df_daily[f"Brent_lag{lag}"] = df_daily["Brent"].shift(lag)
    df_daily[f"iron_lag{lag}"] = df_daily["iron_price"].shift(lag)

for lag in [1,3,5]:
    df_daily[f"coal_lag{lag}"] = df_daily["coal_price"].shift(lag)
    df_daily[f"GPRD_lag{lag}"] = df_daily["GPRD"].shift(lag)

# =========================
# 2. Target 생성
# =========================
df_daily["target"] = df_daily["BDI"].shift(-1)

df_daily = df_daily.dropna().reset_index(drop=True)

# =========================
# 3. 데이터 분리
# =========================
features = [
    "BDI_lag1", "BDI_lag5",
    "Brent_lag1", "Brent_lag5",
    "iron_lag1", "iron_lag5",
    "coal_lag1", "coal_lag5",
    "GPRD_lag1", "GPRD_lag5"
]

X = df_daily[features]
y = df_daily["target"]

split = int(len(df_daily) * 0.8)

X_train = X.iloc[:split]
X_test = X.iloc[split:]

y_train = y.iloc[:split]
y_test = y.iloc[split:]

dates_test = df_daily["Date"].iloc[split:]

# =========================
# 4. 로그 변환
# =========================
y_train_log = np.log1p(y_train)

# =========================
# 5. 모델 정의 (Pipeline 통일)
# =========================
models = {

    # 선형 모델 (스케일링 포함)
    "Linear": Pipeline([
        ("scaler", StandardScaler()),
        ("model", LinearRegression())
    ]),
    
    "Ridge": Pipeline([
        ("scaler", StandardScaler()),
        ("model", Ridge(alpha=1.0))
    ]),
    
    "Lasso": Pipeline([
        ("scaler", StandardScaler()),
        ("model", Lasso(alpha=0.01))
    ]),

    # 트리 모델 (스케일링 필요 없음)
    "RandomForest": RandomForestRegressor(n_estimators=200, random_state=42),
    "ExtraTrees": ExtraTreesRegressor(n_estimators=200, random_state=42),

    "GradientBoosting": GradientBoostingRegressor(random_state=42),
    "HistGradientBoosting": HistGradientBoostingRegressor(random_state=42),

    "XGBOOST": XGBRegressor(random_state=42)
}

# =========================
# 6. 평가 함수
# =========================
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# =========================
# 7. 학습 및 평가
# =========================
results = []
trained_models = {}

for name, model in models.items():
    
    # 학습 (로그 변환 적용)
    model.fit(X_train, y_train_log)
    
    # 예측 (로그 → 원복)
    y_pred_log = model.predict(X_test)
    y_pred = np.expm1(y_pred_log)

    # 평가
    train_score = model.score(X_train, y_train_log)
    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)
    mape_val = mape(y_test, y_pred)

    results.append([name, train_score, r2, rmse, mae, mape_val])
    trained_models[name] = model  

    print(f"\n===== {name} =====")
    print(f"훈련 정확도{train_score:.4f}")
    print(f"R2   : {r2:.4f}")
    print(f"RMSE : {rmse:.4f}")
    print(f"MAE  : {mae:.4f}")
    print(f"MAPE : {mape_val:.2f}%")

    # =========================
    # 시각화
    # =========================
    # plt.figure(figsize=(14,6))
    # plt.plot(dates_test, y_test.values, label="Actual BDI")
    # plt.plot(dates_test, y_pred, label=f"{name} Predicted")
    # plt.legend()
    # plt.title(f"{name} Prediction")
    # plt.xticks(rotation=45)
    # plt.tight_layout()
    # plt.show()

# =========================
# 8. 결과 비교
# =========================
results_df = pd.DataFrame(results, columns=["Model", "train_score","R2", "RMSE", "MAE", "MAPE"])
results_df = results_df.sort_values(by="R2", ascending=False).reset_index(drop=True)

print("\n=== 전체 모델 비교 ===")
print(results_df)

# =========================
# 9. BEST 모델
# =========================
best_model = results_df.iloc[0]

print("\n추천 모델")
print(best_model)
```
